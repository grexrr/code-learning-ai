```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

## Example 1

```python
x = [float(i)/100.0 for i in range(1, 300)]
y = [math.log(i) for i in x]
plt.plot(x, y, 'r-', linewidth=3, label='log_curve')
a = [x[20], x[175]]
b = [y[20], y[175]]
plt.plot(a, b, 'g-', linewidth=2)
plt.plot(a, b, 'b*', markersize=15, alpha=0.75)
plt.legend(loc='upper left')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

## Example 2

```python
u = np.random.uniform(0.0, 1.0, 10000)
plt.hist(u, 80, facecolor='g', alpha=0.75)
plt.grid(True)
plt.show()
```

```python
iteration = 10000
for _ in range(iteration):
    u += np.random.uniform(0.0, 1.0, 10000)

u /= iteration
plt.hist(u, 80, facecolor='g', alpha=0.75)
plt.grid(True)
plt.show()
```

```python
x = np.arange(0.05, 3, 0.05)
y1 = [math.log(num, 1.1)for num in x]
plt.plot(x, y1, linewidth=2, color='#007500', label='log1.1(x)')


y2 = [math.log(num, 2)for num in x]
plt.plot(x, y2, linewidth=2, color='#9f35ff', label='log2(x)')

y3 = [math.log(num, 3)for num in x]
plt.plot(x, y3, linewidth=2, color='#f75000', label='log3(x)')

y4 = [math.log(num, 0.1)for num in x]
plt.plot(x, y4, linewidth=2, color='#535000', label='log0.1(x)')

y5 = [math.log(num, 0.9)for num in x]
plt.plot(x, y5, linewidth=2, color='#C45000', label='log0.9(x)')


plt.plot([1, 1], [y1[0], y5[0]], 'r--', linewidth=2)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
